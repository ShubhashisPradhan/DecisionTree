```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
company=pd.read_csv("/content/sample_data/Company_Data.csv")
company.head()
```

```python
#lets make object column to categorical
for features in company.columns:
  if company[features].dtype=='object':
    company[features]=pd.Categorical(company[features])

company.info()    
```

```python
print("ShelveLoc")
print(company.ShelveLoc.value_counts())
print("******************************")
print("Urban")
print(company.Urban.value_counts())
print("******************************")
print("US")
print(company.US.value_counts())
print("******************************")
```

```python
#lets do label encoding for ShelveLoc and OHE for urban and US
company['ShelveLoc']=company['ShelveLoc'].map({'Medium':1,'Bad':0,'Good':2})
company['Urban']=np.where(company.Urban=='Yes',1,0);
company['US']=np.where(company.US=='Yes',1,0);

company.head()
```

```python
company.info()
```

```python
company.shape
```

lets split the data into test and train

```python
x=company.drop('Sales',axis=1)

x.head()
```

```python
y=company['Sales']
y=pd.cut(y.values,3,labels=['low','medium','high'])
y
```

```python
#y=np.round(y)
#y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

Decision tree building

```python
DT_model=DecisionTreeClassifier(criterion='gini',random_state=1)
DT_model.fit(x_train,y_train)
 
```

```python
for i in range(2,10):
  DT_model=DecisionTreeClassifier(criterion='entropy',max_depth=i,random_state=5)
  DT_model.fit(x_train,y_train)
  print("max_depth = {}".format(i))
  print(DT_model.score(x_train,y_train))
  print(DT_model.score(x_test,y_test)) 
  print("****************************")   
```

```python
DT_model=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=1)
DT_model.fit(x_train,y_train)
print(DT_model.score(x_train,y_train))
print(DT_model.score(x_test,y_test)) 
  
```

```python
#considerinf entropy we are getting good test and train result
DT_model=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=1)
DT_model.fit(x_train,y_train)
print(DT_model.score(x_train,y_train))
print(DT_model.score(x_test,y_test))
```

```python
print(pd.DataFrame(DT_model.feature_importances_,columns=['Importance'],index=x_train.columns))
```

lets try with k-fold valuidation to find our model consistency

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
test_data=company
test_data.copy()
```

```python
x_new=test_data.drop('Sales',axis=1)
x_new=x_new.values
len(x_new)
```

```python
y_new=y
y_new
```

```python
num_fold=5
kfold=KFold(n_splits=num_fold)
model=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=1)
results=cross_val_score(model,x_new,y_new,cv=kfold)
results*100
```
